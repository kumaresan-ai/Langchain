```python
text = [
        "This is a sample text.", 
        "Another sample text.", 
        "Yet another sample text."
        ]
embedded_text = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
vector_store = FAISS.from_texts(text, embedding=embedded_text)
```

```python
vector_store.save_local("faiss_vector_store")
```

```python
retriever = vector_store.as_retriever()
```

```python
query = "sample"
answer = retriever.get_relevant_documents(query)

print(answer)
```

```python
for doc in answer:
    print(doc.page_content)
```
